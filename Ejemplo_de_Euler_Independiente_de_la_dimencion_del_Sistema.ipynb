La siguiente rutina funciona a base de que el objeto en el que guardaremos las soluciones no tiene un tipo de objeto definido en lo absoluto, lo llenamos con un push:

```julia
function euler(f,list,x0)
     x = x0
     h = list[2]-list[1]
     listx=[]
     push!(listx,x)
     for i in 2:length(list)
        t = i*h
        x = x + f(x,t)*h
        push!(listx,x) 
     end
     return listx
end
```

Ejemplo trivial de ecuación diferencial de dos dimensiones (y naturalmente de dos variables) acoplado:

```julia
f(x,t)=[2.5*x[2], -0.1*x[1]^3]
```

Probemos:

```julia
list=0:0.01:20;
vec=euler(f,list,[2,3]);
```

El vector `vec` contiene la solución de nuestro sistema, notese que es un arreglo de objetos `Any` (cualquiera), en este caso vectores de dos dimensiones, noten que la primera entrada es la condición inicial. Una mejora sustancial a este programa puede ser especificar el tipo de las entradas de los vectores que guardaremos en `listx`, trataremos de ver eso despues.

Una forma de extraer una u otra variable es la siguiente (quizá hay una forma más elegante, si la encuentran me dicen):

```julia
listx1=map(x->x[1],vec);
listx2=map(x->x[2],vec);
```
